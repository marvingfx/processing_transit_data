```python
from google.transit import gtfs_realtime_pb2
from datetime import datetime
import requests
import time
import pathlib
import pandas as pd
import os
```

```python
current_path = pathlib.Path().absolute().parent
```

```python
df = pd.DataFrame(columns=['file', 'entities', 'size'])

for index in range(60):
    feed = gtfs_realtime_pb2.FeedMessage()
    response = requests.get('http://gtfs.ovapi.nl/nl/tripUpdates.pb', allow_redirects=True)
    feed.ParseFromString(response.content)
    file_name = '{}/data/GTFS_output_{}_{}.txt'.format(current_path, index, datetime.now().strftime('%Y-%m-%d_%H-%M-%S'))
    
    with open(file_name, 'w+') as file:
        print('writing {} entities to file {}'.format(len(feed.entity), file.name))
        file.write(str(feed))
        df = df.append({'file' : file_name, 'entities' : len(feed.entity), 'size': os.path.getsize(file.name)}, ignore_index=True)
        print(df.tail(1))
             
    time.sleep(60)
    
df.to_csv('{}/data/GTFS_output_df.csv'.format(current_path), index=False)
    
```

```python
df = pd.read_csv('{}/data/GTFS_output_df.csv'.format(current_path))
df.head()
```

```python
def convert_bytes(size):
   for x in ['bytes', 'KB', 'MB', 'GB', 'TB']:
       if size < 1024.0:
           return "%3.1f %s" % (size, x)
       size /= 1024.0

   return size

df['file_size'] = df['size'].map(lambda x : convert_bytes(x))
print(df.head())

print(df[['entities', 'file_size']])

total_size = convert_bytes(df['size'].sum())
min_size = convert_bytes(df['size'].min())
max_size = convert_bytes(df['size'].max())
mean_size = convert_bytes(df['size'].mean())
total_entities = df['entities'].sum()
min_entities = df['entities'].min()
max_entities = df['entities'].max()
mean_entities = df['entities'].mean()
print('total_size: {}'.format(total_size))
print('min_size: {}'.format(min_size))
print('max_size: {}'.format(max_size))
print('mean_size: {}'.format(mean_size))
print('total_entities: {}'.format(total_entities))
print('min_entities: {}'.format(min_entities))
print('max_entities: {}'.format(max_entities))
print('mean_entities: {}'.format(mean_entities))
```
